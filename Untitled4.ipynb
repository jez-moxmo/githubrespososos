```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider, FloatSlider

def plot_sleep_graph(hours_slept=7, sleep_start=22):
    awake_time = sleep_start + hours_slept

    time = np.linspace(0, 40, 40 * 60)
    sleep_cycles = int(hours_slept * 60 / 90)
    stage_percentage = 0.85
    stages = []

    for i in range(sleep_cycles):
        cycle_start = sleep_start + i * 90 / 60
        cycle_end = cycle_start + 1.5

        stage_4_start = cycle_start 
        stage_4_end = cycle_end

        stages.append((stage_4_start, stage_4_end))
        stage_percentage -= 0.1

    brain_waves = np.ones_like(time) * 5

    stage_4_time = 0

    for start, end in stages:
        start_idx = int(start * 60)
        end_idx = int(end * 60)

        stage_4_start_idx = start_idx + int((end_idx - start_idx) * (1 - stage_percentage))
        stage_interval = np.concatenate([
            np.linspace(1, 4, stage_4_start_idx - start_idx),
            np.ones(end_idx - stage_4_start_idx) * 4
        ])
        brain_waves[start_idx:end_idx] = stage_interval[:end_idx-start_idx]
        stage_percentage += 0.1

        stage_4_time += (end_idx - stage_4_start_idx) / 60

    plt.figure(figsize=(12, 6))
    plt.plot(time, brain_waves)
    plt.xlabel('Time (hours)')
    plt.ylabel('Brain Waves (stage)')
    plt.title(f'Single Sleep Cycle - Total Time in Deep Sleep: {stage_4_time:.2f} hours')
    plt.xlim(0, 40)
    plt.ylim(0, 6)
    plt.yticks(range(1, 6), ['Stage 1', 'Stage 2', 'Stage 3', 'Stage 4', 'Awake'])
    plt.xticks(np.arange(0, 41, 1), [f'{i}' for i in range(41)])
    plt.grid()

    plt.axvline(x=awake_time, color='r', linestyle='--', label='Wake up time')
    plt.axvline(x=sleep_start, color='g', linestyle='--', label='Sleep start time')
    plt.axvspan(19, 30, alpha=0.4, color='gray', label='Nighttime')
    plt.legend()

    plt.axvspan(0, 6, alpha=0.4, color='gray')
 
    plt.show()

hours_slept_slider = IntSlider(min=1, max=10, step=1, value=7, description='Hours Slept:')
sleep_start_slider = FloatSlider(min=0, max=23.5, step=0.5, value=22, description='Sleep Start Time:')
interact(plot_sleep_graph, hours_slept=hours_slept_slider, sleep_start=sleep_start_slider)
```
